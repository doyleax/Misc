```python
shows_raw = soup.findAll('td',{'class':'titleColumn'})
shows = []
for x in shows_raw:
    shows.append( x.find('a').text )
shows
```

```python
shows = soup.find('tbody').findAll('a')
urls = []
urls = [x['href'] for x in shows]
urls
```

```python
import pandas as pd
df = pd.DataFrame(urls)
df.rename(columns={0:'urls'},inplace=True)

df['ids'] = df['urls'].str.findall('tt\d.*/')

for row in range(0, len(df)):
    df['ids'][row] = df['ids'][row][0][0:-1]
    
df.drop_duplicates(inplace=True).reset_index(inplace=True)    

df['show'] = shows
```

```python
for X in range(0,len(df)):
    
    
    show_id = df['ids'][X]
    show = df['show'][X]
    cast = 'http://www.imdb.com/title/' + show_id + '/fullcredits?ref_=tt_cl_sm#cast'

    page = requests.get(cast).content
    soup = BeautifulSoup(page,'lxml')

    cast_raw = soup.find('table', {'class':'cast_list'}).findAll('td',{'class':'character'})

    characters = []

    for x in cast_raw:
        try:
            characters.append(x.find('a').text)
        except:
            continue

    characters = pd.DataFrame(characters,columns=['character'])
    characters['show'] = show        
    
    all_characters = all_characters.append(characters)
    
```

```python

all_characters.reset_index(drop=True,inplace=True)
# all_characters.drop('Unnamed: 0',inplace=True)
all_characters.drop(all_characters.columns[0],axis=1,inplace=True)
```

```python
all_characters.character.str.findall('\s(.*)')
```

```python
all_characters['Last'] = all_characters.character.str.findall('\s(.*)')

for x in range(0,len(all_characters)):
    try:
        all_characters['Last'][x] = all_characters['Last'][x][0]
    except:
        all_characters['Last'][x] = ''
```

```python
all_characters['First'] = all_characters.character.str.findall("(.*)\s{1}")
```

```python
for x in range(0,len(all_characters)):
    try:
        all_characters['First'][x] = all_characters['First'][x][0]
    except:
        all_characters['First'][x] = ''
```

```python
# all_characters.First.str.findall("\s{1}")
all_characters.First = all_characters.First.str.strip()
all_characters.First = all_characters.First.str.replace('\s{1}(\B.*\B)','')
```
